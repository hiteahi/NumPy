# Data Types in NumPy

```python
import numpy as np
```

### Check Data Type

```python
arr = np.array([1, 2, 3, 4, 5])
arr.dtype
```

### Change Data Type

```python
l = [1.5, 5.5, 57896.5]
arr = np.array(l, dtype=np.float16)
print(arr.dtype)

arr_int = arr.astype(np.int8)
print(arr_int.dtype)
print(arr_int)
```
